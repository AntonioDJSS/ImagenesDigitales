```python
from skimage import io
from skimage import draw
import numpy as np
import cv2
```

```python
def ImagenCirculos():
  imagen = io.imread("/content/imagen.jpg")
  x,y=draw.circle(200,200,100)
  imagen[x,y]=1

ImagenCirculos()
```

```python
def ImagenRectangulos2(x,y,w,h):
  imagen = io.imread("/content/practica.jpg")
  rr,cc = [x, x + w, x + w, x], [y, y, y + h,y + h]
  return (draw.polygon(rr,cc))

def ImagenRectangulos1():
  imagen = io.imread("/content/practica.jpg")
  rr,cc = ImagenRectangulos2(200,300, 200,300)
  imagen[rr,cc] = 1
  io.imshow(imagen)

ImagenRectangulos1()
```

```python
def TrianguloEquilatero(r1,r2,r3,c1,c2,c3):
   imagen=io.imread("/content/practica.jpg")
   r,c=[r1,r2,r3],[c1,c2,c3]
   rr,cc=draw.polygon(r,c)
   imagen[rr,cc]= 0
   io.imshow(imagen)

TrianguloEquilatero(200,400,400,400,500,300)
```

```python
def TrianguloIsosceles(r1,r2,r3,c1,c2,c3):
   imagen=io.imread("/content/practica.jpg")
   r,c=[r1,r2,r3],[c1,c2,c3]
   rr,cc=draw.polygon(r,c)
   imagen[rr,cc]= 0
   io.imshow(imagen)

TrianguloIsosceles(50,400,400,400,500,300)
```

```python
def TrianguloEscaleno(r1,r2,r3,c1,c2,c3):
   imagen=io.imread("/content/practica.jpg")
   r,c=[r1,r2,r3],[c1,c2,c3]
   rr,cc=draw.polygon(r,c)
   imagen[rr,cc]= 0
   io.imshow(imagen)

TrianguloEscaleno(200,400,400,300,500,300)
```

```python
def ImagenCuadrado2(x,y,w,h):
  imagen = io.imread("/content/practica.jpg")
  rr,cc = [x, x + w, x + w, x], [y, y, y + h,y + h]
  return (draw.polygon(rr,cc))

def ImagenCuadrado1():
  imagen = io.imread("/content/practica.jpg")
  rr,cc = ImagenCuadrado2(200,200, 200,200)
  imagen[rr,cc] = 1
  io.imshow(imagen)

ImagenCuadrado1()
```

```python
def ImagenPoligonoEquilateral(l1,l2,l3,l4,l5,a1,a2,a3,a4,a5):

  imagen=io.imread("/content/imagen.jpg")
  l,a=[l1,l2,l3,l4,l5],[a1,a2,a3,a4,a5]
  ll,aa=draw.polygon(l,a)
  imagen[ll,aa]=0
  io.imshow(imagen)
ImagenPoligonoEquilateral(100,450,450,100,300,30,30,300,300,170)
```
