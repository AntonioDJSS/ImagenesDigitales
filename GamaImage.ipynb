```python
from skimage import exposure;
from skimage import io
from skimage import color
import numpy as np
import matplotlib.pyplot as pp


def CorreccionGama():
  imagen = io.imread("/R.jpg")
  gamma_corrected1 = exposure.adjust_gamma(imagen, 2)
  gamma_corrected2 = exposure.adjust_gamma(imagen, 3)
  pp.figure(0)
  io.imshow(imagen)
  pp.figure(1)
  io.imshow(gamma_corrected1)
  pp.figure(2)
  io.imshow(gamma_corrected2)

print(CorreccionGama())
```
