```python
from skimage import io
from skimage import draw
import numpy as np
import cv2
```

```python
def ImagenCirculos():
  imagen = io.imread("/content/imagen/MOLANG.jpg")
  #draw.circle(cordenadaX,cordenadaY,radio)
  x,y=draw.circle(200,200,100)
  imagen[x,y]=1
  io.imshow(imagen)

ImagenCirculos()
```

```python
def ImagenBezier():
 imagen=io.imread("/content/imagen/molan.jpg")
 #r0,c0 Coordenada unicial, r1,c1 Cordenada media
 #r2,c2 Coordenada ultimo, Coordenada meddia.
 ##draw.bezier_curver(r0,c0,r1,c1,r2,c2,weight)
 x,y=draw.bezier_curve(0,0, 250, 125, 450,500,100)
 #Draw Bezier curve
 imagen[x,y]=1
 io.imshow(imagen)
ImagenBezier()
```

```python
def ImagenRectangulos2(x,y,w,h):
  imagen = io.imread("/content/imagen/piu piu.jpg")
  rr,cc = [x, x + w, x + w, x], [y, y, y + h,y + h]
  return (draw.polygon(rr,cc))

def ImagenRectangulos1():
  imagen = io.imread("/content/imagen/piu piu.jpg")
  rr,cc = ImagenRectangulos2(200,300, 200,300)
  #ImagenRectangulos2(x,y,w,h)
  imagen[rr,cc] = 1
  io.imshow(imagen)

ImagenRectangulos1()
```

```python
def TrianguloEquilatero(r1,r2,r3,c1,c2,c3):
   imagen=io.imread("/content/imagen/piu piu.jpg")
   r,c=[r1,r2,r3],[c1,c2,c3]
   rr,cc=draw.polygon(r,c)
   imagen[rr,cc]= 0
   io.imshow(imagen)

TrianguloEquilatero(200,400,400,400,500,300)
```

```python
from skimage import io
from skimage import draw

def ImagenCuadrado2(x,y,w,h):
  imagen = io.imread("/content/imagen/piu piu.jpg")
  rr,cc = [x, x + w, x + w, x], [y, y, y + h,y + h]
  return (draw.polygon(rr,cc))

def ImagenCuadrado1():
  imagen = io.imread("/content/imagen/piu piu.jpg")
  rr,cc = ImagenCuadrado2(200,200, 200,200)
  #ImagenCuadrado2(x,y,w,h)
  imagen[rr,cc] = 1
  io.imshow(imagen)

ImagenCuadrado1()
```

```python
def ImagenPoligonoEquilateral(l1,l2,l3,l4,l5,a1,a2,a3,a4,a5):

  imagen=io.imread("/content/imagen/piu piu.jpg")
  l,a=[l1,l2,l3,l4,l5],[a1,a2,a3,a4,a5]
  ll,aa=draw.polygon(l,a)
  imagen[ll,aa]=0
  io.imshow(imagen)
ImagenPoligonoEquilateral(100,450,450,100,300,30,30,300,300,170)
```
