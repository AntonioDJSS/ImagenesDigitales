**`LIBRERIAS`**

```python
from skimage import io
from skimage import color
import numpy as np
import matplotlib.pyplot as pp
```

**`CONVERSOR DE COLORES`**

```python
imagen = io.imread('/content/Imagen.jpg')

def ConversionHSV(imagen):
    # De RGB a HSV
    lab = color.rgb2hsv(imagen)
    # De HSV a RGB
    rgb = color.hsv2rgb(lab)

    pp.figure(0)
    io.imshow(lab)
    pp.figure(1)
    io.imshow(rgb)

print(ConversionHSV(imagen))
```

```python
def ConversionXYZ(imagen):
  #De RGB a XYZ
  xyz = color.rgb2xyz(imagen)
  #De XYZ a RGB
  rgb = color.xyz2rgb(xyz)

  pp.figure(0)
  io.imshow(xyz)
  pp.figure(1)
  io.imshow(rgb)

print(ConversionXYZ(imagen))
```

```python
def ConversionYiq(imagen):
  #De RGB a Yiq
  yiq = color.rgb2yiq(imagen)
  #De Yiq a RGB
  rgb = color.yiq2rgb(yiq)

  pp.figure(0)
  io.imshow(yiq)
  pp.figure(1)
  io.imshow(rgb)

print(ConversionYiq(imagen))
```

```python
def ConversionYPbPr(imagen):
  #De RGB a YPbPr
  YPbPr = color.rgb2ypbpr(imagen)
  #De YPbPr a RGB
  rgb = color.ypbpr2rgb(YPbPr)

  pp.figure(0)
  io.imshow(YPbPr)
  pp.figure(1)
  io.imshow(rgb)

print(ConversionYPbPr(imagen))
```

```python
def ConversionYuv(imagen):
  #De RGB a Yuv
  yuv = color.rgb2yuv(imagen)
  #De Yub a RGB
  rgb = color.yuv2rgb(yuv)

  pp.figure(0)
  io.imshow(yuv)
  pp.figure(1)
  io.imshow(rgb)

print(ConversionYuv(imagen))
```

**`CANALES DE COLOR`**

```python
#RED = 0
#GREEN = 1
#BLUE = 2
def ColorVerde():
  verde = np.copy(imagen)
  verde[:,:,0] = 0
  verde[:,:,2] = 0
  pp.title("Canal Verde")
  pp.imshow(verde)

print(ColorVerde())
```

```python
def ColorAzul():
  azul = np.copy(imagen)
  azul[:,:,0] = 0 #RED
  azul[:,:,1] = 0 #GREEN
  pp.title("Canal Azul")
  pp.imshow(azul)

print(ColorAzul())
```

```python
def ColorCafe():
  #xyz = color.rgb2xyz(imagen)
  cafe = np.copy(imagen)
  cafe[:,:,2] = 0
  pp.title("Canal Cafe")
  pp.imshow(cafe)

print(ColorCafe())
```
