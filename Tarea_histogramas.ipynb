```python
from matplotlib import pyplot as pp
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/contraste.png')
img1 = cv2.imread('/content/sin_contraste.jpg')

def histogramaColores(imagenes):
  cv2_imshow(imagenes)
  color = ('b','g','r')
  for i , c in enumerate(color):
    hist = cv2.calcHist([imagenes], [i], None, [256], [0,256])
    pp.plot(hist, color = c)
    pp.xlim([0,256])
  pp.show()


def run():
  print('IMAGEN CON CONTRASTE')
  histogramaColores(img)
  print('\n\nIMAGEN SIN CONTRASTE')
  histogramaColores(img1)


if __name__ == '__main__':
  run()
```

```python
from matplotlib import pyplot as pp
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
import numpy as np


img = cv2.imread('/content/contraste.png')
img1 = cv2.imread('/content/sin_contraste.jpg')  


def histogramaColorDer(imagenes):
  pixeles = imagenes.shape
  alto = pixeles[0]
  ancho = pixeles[1]
  ancho = int(ancho/2)
  mask = np.zeros(imagenes.shape[:2], np.uint8)
  mask[0:alto, 0:ancho] = 255
  masked_img = cv2.bitwise_and(imagenes, imagenes, mask = mask)
  color = ('b','g','r')
  for i , c in enumerate(color):
    hist = cv2.calcHist([imagenes], [i], mask, [256], [0,256])
    pp.plot(hist, color = c)
    pp.subplot(221), pp.imshow(imagenes)
    pp.subplot(222), pp.imshow(mask)
    pp.subplot(223), pp.imshow(masked_img)
    pp.subplot(224), pp.plot(hist)
    pp.xlim([0,256])
  print(f'Resolucion de la imagen:\n Alto = {alto}px \n Ancho = {ancho}px')
  pp.show()


def histogramaColorIzq(imagenes):
  pixeles = imagenes.shape
  alto = pixeles[0]
  ancho = pixeles[1]
  ancho = int(ancho/2)
  mask = np.zeros(imagenes.shape[:2], np.uint8)
  mask[0:alto, ancho:700] = 255
  masked_img = cv2.bitwise_and(imagenes, imagenes, mask = mask)
  color = ('b','g','r')
  for i , c in enumerate(color):
    hist = cv2.calcHist([imagenes], [i], mask, [256], [0,256])
    pp.plot(hist, color = c)
    pp.subplot(221), pp.imshow(imagenes)
    pp.subplot(222), pp.imshow(mask)
    pp.subplot(223), pp.imshow(masked_img)
    pp.subplot(224), pp.plot(hist)
    pp.xlim([0,256])
  print(f'Resolucion de la imagen:\n Alto = {alto}px \n Ancho = {ancho}px')
  pp.show()


def run():
  print('IMAGEN CON CONTRASTE')
  histogramaColorDer(img)
  print('\n\nIMAGEN SIN CONTRASTE')
  histogramaColorDer(img1)
  print('IMAGEN CON CONTRASTE IZQUIERO')
  histogramaColorIzq(img)
  print('\n\nIMAGEN SIN IZQUIERO')
  histogramaColorIzq(img1)

if __name__ == '__main__':
  run()
```
