```python
from skimage import io
from skimage import color
from pylab import *
import matplotlib.pyplot as pp

def conversion(imagen):
    #De rgb a lab 
    imagen_lab = color.rgb2lab(Imagen)
    #De lab a rgb
    imagen_rgb = color.lab2rgb(imagen_lab)
    
    pp.figure(0)
    io.imshow(imagen_lab)
    pp.figure(1)
    io.imshow(imagen_rgb)
    
Imagen = io.imread('/content/minion.jpg')
conversion(Imagen)
```
