```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import io

imagen = cv2.imread('/content/F100011811.jpg')

cv2.circle(imagen, (20, 30), 7, (255, 0, 0), 2)
cv2.circle(imagen, (540, 80), 7, (0, 255, 0), 2)
cv2.circle(imagen, (80, 300), 7, (0, 0, 255), 2)
cv2.circle(imagen, (500, 200), 7, (255, 255, 0), 2)

"""
Calcula una tranformacion de perspectiva a partir de cuatro pares de los puntos correspondientes.
La funcion calcula la matriz 3x3 de una tranformacion de perspectiva de modo que:
"""

pts1 = np.float32([[20,30], [540,80], [80,300], [500,200]])
pts2 = np.float32([[0,0], [624,0], [0,385], [624,385]])

M = cv2.getPerspectiveTransform(pts1, pts2)
dst = cv2.warpPerspective(imagen, M, (624,385))

cv2_imshow(imagen)
cv2_imshow(dst)

```
