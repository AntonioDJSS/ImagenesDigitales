```python
from pylab import *
import matplotlib.pyplot as pp
from skimage import exposure
from skimage import io
import cv2
```

```python
imagen = cv2.imread("/content/rog.jpg")

letra = int(input('¿Que tipo de letra quiere utilizar? \n1) COMPLEX SMALL \n2) COMPLEX \n3) SCRIPT SIMPLEX \n4) ITALIC \nTIPO DE LETRA ELEGIDA: '))

r = int(input('\nGenerador de Colores: \nInserta color de r: '))
g = int(input('Inserta color de g: '))
b = int(input('Inserta color b: '))

print('\nCoordenadas: ', imagen.shape)
x = int(input('\nGenerador de Coordenadas: \nIngresa valor de X: '))
y = int(input('Ingresa valor de y: '))

def TextoImagenes():

  if letra == 1:
    cv2.FONT_HERSHEY_COMPLEX_SMALL
  elif letra == 2:
    cv2.FONT_HERSHEY_COMPLEX
  elif letra == 3:
    cv2.FONT_HERSHEY_SCRIPT_COMPLEX
  elif letra == 4: 
    cv2.FONT_ITALIC   

  cv2.putText(imagen, "Hola Sistemas", (x,y), letra, 4.8,(r,g,b),5,cv2.LINE_AA)
  io.imshow(imagen)

print(TextoImagenes())
```
