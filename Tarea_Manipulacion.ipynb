```python
#PUNTO 1
from skimage import exposure
from skimage import io
from pylab import *
import numpy as np
import matplotlib.pyplot as pp
def CorrectionGama():

  imagen = io.imread("/content/carbon.png")
  gama_corrected1 = exposure.adjust_gamma(imagen, 5)
  gama_corrected2 = exposure.adjust_gamma(imagen, 10)
  pp.figure(0)
  io.imshow(imagen)
  pp.figure(1)
  io.imshow(gama_corrected1)
  pp.figure(2)
  io.imshow(gama_corrected2)
CorrectionGama()
```

```python
#PUNTO 2
from skimage import io
from pylab import *
from skimage.transform import resize
def TamañoImagen():
  imagen = io.imread("/content/carbon.png")
  img_res = resize(imagen, (100,100))
  io.imshow(img_res)
  io.imsave("ImagenNueva.png", img_res)
TamañoImagen()
```

```python
#PUNTO 3
from skimage import exposure
from skimage import io
from pylab import *
import numpy as np
import matplotlib.pyplot as pp
from skimage.transform import rotate
imagen =io.imread("/content/carbon.png")
x = int(input(' coloca la cordenada de x: \n'))
def RotacionImagen():
  
  rotacionImagen = rotate(imagen, x)
  print('IMAGEN ORIGINAL')
  pp.figure(0)
  io.imshow(imagen)
  print('IMAGEN ROTADA')
  pp.figure(1)
  io.imshow(rotacionImagen)

RotacionImagen()
```

```python
from pylab import *
import matplotlib.pyplot as pp
from skimage import exposure
from skimage import io
import cv2

imagen = cv2.imread("/content/carbon.png")

letra = int(input('\n¿Que tipo de letra quiere utilizar? \n1) COMPLEX SMALL \n2) COMPLEX \n3) SCRIPT SIMPLEX \n4) ITALIC \nTIPO DE LETRA ELEGIDA: '))

color = int(input('\n¿Que color quiere utilizar? \n1) NEGRO  \n2) BLANCO  5\n3) AZUL REY  \n4) ROJO \nTIPO DE COLOR ELEGIDO: '))

texto1= input('\nIntroduce tu texto: ')

if color == 1:
  r = int(0)
  g = int(0)
  b = int(0)
elif color == 2:
  r = int(255)
  g = int(255)
  b = int(255)
elif color == 3:
  r = int(0)
  g = int(0)
  b = int(255)
elif color == 4:
  r = int(255) 
  g = int(0)
  b = int(26)

print('\nCoordenadas: ', imagen.shape)
x = int(input('\nGenerador de Coordenadas: \nIngresa valor de X: '))
y = int(input('Ingresa valor de y: '))

ancho = imagen.shape[0]
alto = imagen.shape[1]

if x > ancho or y > alto:
  print('\nTe saliste del rango')
else:
  if letra == 1:
    cv2.FONT_HERSHEY_COMPLEX_SMALL
  elif letra == 2:
    cv2.FONT_HERSHEY_COMPLEX
  elif letra == 3:
    cv2.FONT_HERSHEY_SCRIPT_COMPLEX
  elif letra == 4: 
    cv2.FONT_ITALIC   

  cv2.putText(imagen, texto1, (x,y), letra, 4.8,(r,g,b),5,cv2.LINE_AA)
  io.imshow(imagen)
```
